```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

df = pd.read_csv('USCensus1990-2000rows.data.txt')
df
```

```python
data = df.iloc[:,1:]
data
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram

Z = linkage(data, 'ward')   
# Z is now the clustered data
# Graph the data using matplotlibplt.figure()
dendrogram(Z)
plt.show()
```
